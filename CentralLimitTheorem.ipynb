# The Central Limit Theorem

```python
import numpy as np
from matplotlib import pyplot as plt
np.random.randint(0,4,100)
import seaborn as sns
```

```python
# take from uniform distribution
dist = []
i=0
while i < 20:
    dist.append( np.random.uniform(0,1000, 1000).sum()**2 )
    i+=1
    
```

```python
sns.distplot(dist)
plt.show()
```
