```python
import pandas as pd
import numpy as np
from nltk.corpus import stopwords
import string
```

```python
df = pd.read_csv('labeledTrainData.tsv', sep='\t')
```

```python
df.head()
```

```python
from bs4 import BeautifulSoup
```

```python
def rem_html(text):
    return BeautifulSoup(text, 'html.parser').get_text()
```

```python
def rem_punc(text):
    nopunc = [char for char in text if char not in string.punctuation]
    return ''.join(nopunc)
```

```python
def rem_stop(text):
    return [word for word in text.split() if word.lower() not in stopwords.words('english')]
```

```python
df['review_cleaned'] = df['review'].apply(rem_html)
```

```python
df['review_cleaned2'] = df['review_cleaned'].apply(rem_punc)
```

```python
df['review_cleaned3'] = df['review_cleaned2'].apply(rem_stop)
```

```python
from nltk.stem.snowball import SnowballStemmer
```

```python
stemmer = SnowballStemmer("english")
```

```python
print(stemmer.stem("running"))
```

```python
def stem(text):
    return [stemmer.stem(element) for element in text]
```

```python
df['review_cleaned4'] = df['review_cleaned3'].apply(stem)
```

```python
df
```

```python
from nltk.stem.wordnet import WordNetLemmatizer
```

```python
lmtzr = WordNetLemmatizer()
```

```python
lmtzr.lemmatize('praised', 'v')
```

```python
def lemma(text):
    return [lmtzr.lemmatize(element, 'v') for element in text]
```

```python
df['review_cleaned5'] = df['review_cleaned3'].apply(lemma)
```

```python
df
```

```python
def lemma_noun(text):
    return [lmtzr.lemmatize(element, 'n') for element in text]
```

```python
df['review_cleaned6'] = df['review_cleaned5'].apply(lemma_noun)
```
